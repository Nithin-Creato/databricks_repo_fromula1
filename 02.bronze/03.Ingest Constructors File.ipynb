```python
source_file = f"{landing_folder_path}/constructors.json"
table_name = f"{catalog_nameone}.{bronze_schema}.constructors"
```

```python
construct_schema = """constructorId STRING, 
                      name STRING, \
                      nationality STRING, 
                      url STRING"""

constructor_df = (
    spark.read.format("json")
    .schema(construct_schema)
    .option("mode", "FAILFAST")
    .load(source_file)
)

constructor_final_df = add_ingestion_metadata(constructor_df)
display(constructor_final_df)
```

```python
constructor_final_df.write.format('delta').mode('overwrite').saveAsTable(table_name)
```
