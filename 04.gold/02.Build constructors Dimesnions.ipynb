```python
from pyspark.sql import functions as F
```

```python
target_table = f"{catalog_nameone}.{gold_schema}.dim_constructors"
```

```python
constructors_df = spark.table(f"{catalog_nameone}.{silver_schema}.constructors")
```

```python
dim_constructors_df = (
    constructors_df
    .select(
        "constructor_id",
        "constructor_name",
        "nationality"
        )
)
```

```python
(
    dim_constructors_df
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(target_table)
)
```
