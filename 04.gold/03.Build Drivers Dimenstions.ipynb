```python
from pyspark.sql import functions as F
```

```python
target_table = f"{catalog_nameone}.{gold_schema}.dim_drivers"
```

```python
drivers_df = spark.table(f"{catalog_nameone}.{silver_schema}.drivers")
```

```python
dim_drivers_df = (
    drivers_df
    .select(
        "driver_id",
        "driver_name",
        "date_Of_Birth",
        "nationality"
        )
)
```

```python
(
    dim_drivers_df
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(target_table)
)
```
