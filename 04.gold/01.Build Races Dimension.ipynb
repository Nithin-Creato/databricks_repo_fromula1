1. Read silver races table 
2. Read silver circuits table
3. join the date from races with circuits USING circuits_id
4. Select required columns
5. write transformed data to gold

```python
from pyspark.sql import functions as F
```

```python
target_table = f"{catalog_nameone}.{gold_schema}.circuits"
```

```python
circuits_df = spark.table(f"{catalog_nameone}.{silver_schema}.circuits")
races_df = spark.table(f"{catalog_nameone}.{silver_schema}.races")
```

```python
dim_races_df = (
    races_df
        .join(
            circuits_df,
            races_df.circuit_id == circuits_df.circuit_id,
            "inner" 
        )
        .select(
            races_df.season,
            races_df.round,
            races_df.race_name,
            races_df.race_date,
            circuits_df.circuit_name,
            circuits_df.locality,
            circuits_df.country
        )
)

display(dim_races_df)
```

```python
(
dim_races_df
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(target_table)
)
```
